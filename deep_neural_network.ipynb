```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
np.random.seed(0)
```

```python
"""In this experiment, we will use only 1 dense hidden layer with 4 neuron to train a simple neuron network"""
```

```python
n_pts = 500
X, y = datasets.make_circles(n_samples= n_pts, 
                             random_state= 123, 
                             noise=0.1, 
                             factor=0.2)
```

```python
"""A high noise (0.8) will cause the data points to become overly convoluted, making it a lot harder to classify the data. You want to ensure that the noice is kept relatively low (0.1)"""
```

```python
print(X)
print(y)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation = 'sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Dense( number of neurons, input_shape=(, ,), activation='sigmoid')
```

```python
model.compile(Adam(lr= 0.01), 'binary_crossentropy', metrics=['accuracy'])
```

```python
his = model.fit(x=X, y=y, 
          verbose= 1, 
          batch_size = 20, 
          epochs = 100,
         shuffle = 'true')
```

```python
plt.plot(his.history['accuracy'])
plt.xlabel('epoch')
plt.legend(['accuracy'])
plt.title('accuracy')
```

```python
plt.plot(his.history['loss'])
plt.xlabel('epoch')
plt.legend(['loss'])
plt.title('loss')
```

```python
"""Run the cell and looking at this it seems evident that the full 100 IPAC run through was required for

the last function to fully minimize.

This is actually indicative of the fact that our dataset was more difficult to classify as opposed to

the linear separable data that we dealt with in the last section."""
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 0.25, max(X[:,0]) + 0.25, 50)
    y_span = np.linspace(min(X[:,1]) - 0.25, max(X[:,1]) + 0.25, 50)
    xx, yy = np.meshgrid(x_span, y_span)
    xx_, yy_ = xx.ravel(), yy.ravel()
    grid = np.c_[xx_, yy_]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
 
```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
#-----------------------------------

x = 0.1
y = 0.0
point = np.array([[x, y]])
prediction = model.predict(point)
plt.plot([x], [y], marker = 'o', markersize=10, color="red")
print("Prediction is : ", prediction)
```
